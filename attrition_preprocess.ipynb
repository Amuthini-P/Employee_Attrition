```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import mode
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
df=pd.read_csv("employee_attrition.csv")
```

```python
# The original Dataset:
df.head()
```

```python
# Label Encoding the Attrition , BusinessTravel , Gender and JobRole Column

import pandas as pd
from sklearn.preprocessing import LabelEncoder

data = df
df1 = pd.DataFrame(data)

# Columns to encode
columns_to_encode = ['Attrition', 'BusinessTravel','Gender']

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Iterate over each column and encode its values
for column in columns_to_encode:
    df1[column] = label_encoder.fit_transform(df1[column])

print(df1['Attrition'])
print(df1['BusinessTravel'])
print(df1['Gender'])
```

```python
x = df1[['Age','BusinessTravel','DailyRate','DistanceFromHome','Gender','JobSatisfaction','MonthlyRate','RelationshipSatisfaction','TotalWorkingYears','WorkLifeBalance']]
y = df1['Attrition']
print(x)
```

```python
mean_x = np.mean(x)
mean_y = np.mean(y)

# Median
median_x = np.median(x)
median_y = np.median(y)

# Mode
mode_x = mode(x)
mode_y = mode(y)

print("\nCentral Tendency Measures:")
print("Mean of x:", mean_x)
print("Mean of y:", mean_y)
print("Median of x:", median_x)
print("Median of y:", median_y)
print("Mode of x:", mode_x)
print("Mode of y:", mode_y)
```

```python
# Range
range_x = np.ptp(x)
range_y = np.ptp(y)

# Variance
variance_x = np.var(x)
variance_y = np.var(y)

# Standard Deviation
std_dev_x = np.std(x)
std_dev_y = np.std(y)

# Interquartile Range
interquartile_range_x = np.percentile(x, 75) - np.percentile(x, 25)
interquartile_range_y = np.percentile(y, 75) - np.percentile(y, 25)

print("\nVariability Measures:")
print("Range of x:", range_x)
print("Range of y:", range_y)
print("\n\nVariance of x:\n", variance_x)
print("\n\nVariance of y:\n", variance_y)
print("\n\nStandard Deviation of x:\n", std_dev_x)
print("\n\nStandard Deviation of y:\n", std_dev_y)
print("\n\nInterquartile Range of x:", interquartile_range_x)
print("\n\nInterquartile Range of y:", interquartile_range_y)
```

```python
from scipy.stats import skew, kurtosis

# Skewness
skewness_x = skew(x)
skewness_y = skew(y)

# Kurtosis
kurtosis_x = kurtosis(x)
kurtosis_y = kurtosis(y)

print("\nShape Measures:")
print("\nSkewness of x:\n", skewness_x)
print("\nSkewness of y:\n", skewness_y)
print("\nKurtosis of x:\n", kurtosis_x)
print("\nKurtosis of y:\n", kurtosis_y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

# Monthly rate data
monthly_rate_data = df1['MonthlyRate']

# Calculate skewness
skewness = skew(monthly_rate_data)

# Plot histogram
plt.figure(figsize=(12, 4))

# Positive skewness
plt.subplot(1, 3, 1)
plt.hist(monthly_rate_data, bins=10, color='skyblue', edgecolor='black')
plt.title("Positive Skewness")
plt.xlabel("Monthly Rate")
plt.ylabel("Frequency")

# Negative skewness
plt.subplot(1, 3, 2)
plt.hist(100 - monthly_rate_data, bins=10, color='red', edgecolor='black')
plt.title("Negative Skewness")
plt.xlabel("Monthly Rate")
plt.ylabel("Frequency")

# Normal skewness
plt.subplot(1, 3, 3)
plt.hist(np.random.permutation(monthly_rate_data), bins=10, color='blue', edgecolor='black')
plt.title("Normal Skewness")
plt.xlabel("Monthly Rate")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

# Print skewness values
print("\nSkewness Values:")
print("Positive Skewness:", skewness)
print("Negative Skewness:", -skewness)
print("Normal Skewness:", skew(np.random.permutation(monthly_rate_data)))
```

```python
import numpy as np
import pandas as pd


data = df1

missing_indices = np.random.choice(df1.index, size=5, replace=False)
missing_columns = np.random.choice(df1.columns, size=2, replace=False)
df1.loc[missing_indices, missing_columns] = np.nan
print("Dataset with Missing Values:")
print(df1)
```

```python
df_listwise = df1.dropna()

print("After Listwise Deletion:")
print(df_listwise)
```

```python
df_pairwise = df1.dropna(axis=0, how='any')

print("After Pairwise Deletion:")
print(df_pairwise)
```

```python
# Dropping variables (dropping columns with missing values)
df_dropped = df1.dropna(axis=1)

print("After Dropping Variables:")
print(df_dropped)
```

```python
# Imputation: Mean
df_mean_imputed = df1.fillna(df1.mean())

print("After Mean Imputation:")
print(df_mean_imputed)

# Imputation: Median
df_median_imputed = df1.fillna(df1.median())

print("After Median Imputation:")
print(df_median_imputed)

# Imputation: Mode
df_mode_imputed = df1.apply(lambda x: x.fillna(x.mode()[0]))

print("After Mode Imputation:")
print(df_mode_imputed)
```
